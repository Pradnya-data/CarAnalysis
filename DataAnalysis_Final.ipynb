# Treating Missing Values

```python
# seggregate columns categorical and continuous values

categorical=[]
continuous=[]

for i in df.columns:
    if df[i].dtype=="object":
        categorical.append(i)
    else:
        continuous.append(i)
        
```

```python
categorical
```

```python
continuous
```

```python
# for numerical column replace null values with mean/median
# for categorical column repace null values with mode
df['Price'].mean()
```

```python
df['Type'].mode()
```

```python
# check modal value of all categorical columns
for i in categorical:
    print(f"Modal value of {i} is {df[i].mode()[0]}")
```

```python
for i in continuous:
    print(i,df[i].mean())
```

```python
# fill up missing values fillna()

for i in df:
    if df[i].dtypes=="object":
        m=df[i].mode()[0]
        df[i]=df[i].fillna(m)
    else:
        me=df[i].mean()
        df[i]=df[i].fillna(me)
```
